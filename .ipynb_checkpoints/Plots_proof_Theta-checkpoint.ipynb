```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# time_data_points= pd.read_pickle("./heap_sort_google_collab.txt")

time_data_points= dict ()
f=open('merge_sort_02.txt', 'r')
time_data_points= eval( f.read())
time_data_points
```

```python
from scipy import interpolate
# Extract the keys and values and plot them

keys, values = zip(*time_data_points.items()) 
size_n =    np.asarray(keys)   # np.log2( np.asarray(keys) ) *
time_y =  np.asarray(values) *1000 # microseconds

n_log_n_large= 0.05*np.asarray(keys) * np.log2( np.asarray(keys) )
n_log_n_small= 0.03*np.asarray(keys) * np.log2( np.asarray(keys) )

original_data = pd.DataFrame({  "x":size_n, "y":time_y  })

plt.figure(figsize=(10,10))
# plt.plot( original_data['x'],  original_data['y'])
plt.scatter( original_data['x'],  original_data['y'],marker="+")
plt.plot( original_data['x'],  original_data['y'])
plt.plot( n_log_n_large,  original_data['y'] , color='red')
plt.plot( n_log_n_small,  original_data['y'] , color='black')
plt.show()

```
