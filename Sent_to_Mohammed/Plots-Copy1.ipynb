```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pandasql as ps

read= pd.read_pickle("./data_frame.txt")
read
```

```python
# Extract the keys and values and plot them

min_query = """SELECT Sort_Type, 
Data_Type_or_List_type,
List_length, 
min( Runtimes/Number_of_repeatitions)  as Single_runtime
FROM read 
GROUP BY Sort_Type, 
Data_Type_or_List_type,
List_length """

df_min = pd.DataFrame( ps.sqldf(min_query) )
df_min

```

```python
df_min.columns
```

```python
filter = ( df_min['Sort_Type']=='<built-in function sorted>') &  (df_min['Data_Type_or_List_type']=='random')
plot_this=df_min[filter]
plot_this
```

```python

plt.figure(figsize=(25,10))
plt.plot ( plot_this['List_length'],  plot_this['Single_runtime'])
plt.scatter  ( plot_this['List_length'],  plot_this['Single_runtime'], color='black')
plt.show()
```

```python
plot_lower_half=plot_this[(plot_this['List_length']<=10240)]
plot_lower_half
```

```python
plt.figure(figsize=(25,10))
plt.plot ( plot_lower_half['List_length'],  plot_lower_half['Single_runtime'])
plt.scatter  ( plot_lower_half['List_length'],  plot_lower_half['Single_runtime'], color='black')
plt.show()
```
